### User IDs

### User Watching History

```python
relation = pd.read_csv("relation.csv")
relation
```

```python
relation["user_id"] = relation.user_id.astype("string")
relation.info()
```

```python
s = []
e = []
d = []
for x in relation["start_time"]:
    num = list(map(str, x.split(":")))
    num[0] = int(num[0])*3600
    num[1] = int(num[1])*60
    num[2] = int(num[2])
    s.append(sum(num))
for x in relation["end_time"]:
    num = list(map(str, x.split(":")))
    num[0] = int(num[0])*3600
    num[1] = int(num[1])*60
    num[2] = int(num[2])
    e.append(sum(num))
for i, x in enumerate(s):
    if e[i]-x < 0:
        d.append(86400+e[i]-x)
    else:
        d.append(e[i]-x)


relation["s"] = s
relation["e"] = e
relation["d"] = d
```

### % of watched content

```python

watching = list((relation["d"]*100)/relation["durata"])
relation["watching%"] = watching
new_relation = relation.drop(["date","start_time","end_time","s","e","durata","d"], axis = 1)
new_relation
```

### Users Watched History 

```python
di = {}
for i,x in enumerate(new_relation["user_id"]):
    if x in di:
        di[x].append([new_relation["watching%"][i], new_relation["content_id"][i]])
    else:
        di[x] = [[new_relation["watching%"][i], new_relation["content_id"][i]]]

# sort the value of dictionary items.....

for x in di:
    di[x].sort(reverse=True)
di
```

### Content Details

```python
cntnt = pd.read_csv("content.csv")
cntnt
```

```python
cntnt["language"].value_counts()
```

### Arrange the columns

```python
cntnt = cntnt[["content_id","content_type","language","genre","duration","release_date","rating"]]
cntnt
```

### Convert Data Type

```python
cntnt["c"] = cntnt.duration.astype("string")
cntnt["a"] = cntnt.release_date.astype("string")
cntnt["b"] = cntnt.rating.astype("string")

cntnt.drop(["duration","release_date","rating"], axis = 1, inplace = True)
cntnt.head(5)
```

### Combine all columns in single column

```python
cntnt["tag"] = cntnt["content_type"]+" "+cntnt["language"]+" "+cntnt["genre"]+" "+cntnt["c"]+" "+cntnt["a"]+" "+cntnt["b"]
cntnt.drop(["content_type", "language", "genre", "c", "a", "b"], axis = 1, inplace = True)
cntnt.head(5)
```

### For Stem the element of "tag"

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

cntnt["tag"] = cntnt["tag"].apply(stem)
```

```python
cntnt.sample(10)
```

### Vectorization of "tag"

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 15000, stop_words = "english")
```

```python
vector = cv.fit_transform(cntnt["tag"]).toarray()

vector[0]
```

```python
cv.get_feature_names()[-10:]
```

### Cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
cs1 = cosine_similarity(vector)

cs1
```

### Recommendation

```python
def recommended(m_id):
    content_index = cntnt[cntnt["content_id"] == m_id].index[0]
    distances = cs1[content_index]
    
    recommended_content = list(sorted(list(enumerate(distances)), reverse=True, key= lambda x: x[1])[1:21])
    
    cont = []
    for x in recommended_content:
        cont.append(cntnt["content_id"][x[0]])
    
    rc = []
    for x in recommended_content:
        rc.append((x[1]))
    c_cont = dict(zip(cont, rc))
    
    return c_cont

def show(x):
    ans = {}
    try:
        for i in range(5):
            ans.update(recommended(di[x][:][i][1]))
    except:
        for i in range(len(di[x][:])):
            ans.update(recommended(di[x][:][i][1]))
    ddd = dict(sorted(ans.items(), key=lambda y: y[1]))
    give = ddd.keys()
    for x in give:
        print(x)

for x in test_data["user_id"]:
    print("_________________________________________________{}______________________________________________".format(x))
    show(x)
```
